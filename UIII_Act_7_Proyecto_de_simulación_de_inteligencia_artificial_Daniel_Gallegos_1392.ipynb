```python
# Simulador de Asistente Dental con Inteligencia Artificial Básica
print("====================================================")
print("¡Bienvenido a Clinica Dental 'Sonrisas Lindas'!")
print("Iniciando DentiBot - Tu asistente virtual con IA...")
print("====================================================")

# Pedimos el nombre del paciente para personalizar la experiencia
nombre_paciente = input("Para comenzar, ¿cual es tu nombre? ")
print("\nHola " + nombre_paciente + ", soy DentiBot. Hoy te ayudare.")

# Variable de control para mantener el menú activo
continuar = "si"

# Ciclo principal del programa (Menú interactivo)
while continuar == "si":
    print("\n--- MENU DE ATENCION ---")
    print("1. Tengo un dolor o emergencia")
    print("2. Quiero estetica (limpieza o blanqueamiento)")
    print("3. Revisar horarios y costos")
    print("4. Salir")

    opcion = input("\nPor favor, elige el numero de tu opcion: ")

    # Evaluar la opción del usuario (Simulación de toma de decisiones de la IA)
    if opcion == "1":
        print("\n[DentiBot analiza los sintomas...]")
        print("Lamento escuchar eso, " + nombre_paciente + ".")
        tipo_dolor = input("¿El dolor es 'fuerte' o 'leve'?: ")

        if tipo_dolor == "fuerte":
            print("-> Alerta: Puede ser una infeccion o una caries profunda.")
            print("-> Recomendacion: Te asignamos una cita de URGENCIA para hoy mismo.")
        elif tipo_dolor == "leve":
            print("-> Diagnostico preliminar: Sensibilidad dental o caries inicial.")
            print("-> Recomendacion: Agenda una cita general esta semana y evita cosas muy frias.")
        else:
            print("-> No reconozco ese tipo de dolor, pero te sugerimos venir a revision.")

    elif opcion == "2":
        print("\n[DentiBot revisa opciones de estetica...]")
        print("¡Excelente eleccion para cuidar tu sonrisa!")
        servicio = input("¿Prefieres 'limpieza' o 'blanqueamiento'?: ")

        if servicio == "limpieza":
            print("-> La limpieza elimina el sarro. Se recomienda cada 6 meses.")
            print("-> ¿Deseas agendar? (Escribe al final en el menu principal).")
        elif servicio == "blanqueamiento":
            print("-> El blanqueamiento aclara tus dientes. Requiere una revision previa de tus encias.")
        else:
            print("-> Por ahora solo contamos con limpieza y blanqueamiento en esta seccion.")

    elif opcion == "3":
        print("\n--- INFORMACION GENERAL ---")
        print("-> Horarios: Lunes a Viernes de 9:00 AM a 6:00 PM.")
        print("-> Costo de Consulta General: $300 pesos.")
        print("-> Aceptamos efectivo y tarjetas de credito.")

    elif opcion == "4":
        print("\nGracias por usar DentiBot, " + nombre_paciente + ". ¡Que tengas un gran dia!")
        continuar = "no" # Esto rompe el ciclo while y cierra el programa

    else:
        print("\n[Error de lectura] Opcion no valida. Por favor, selecciona 1, 2, 3 o 4.")

    # Si el usuario no eligió salir, preguntamos si quiere hacer algo más
    if opcion != "4":
        continuar = input("\n¿Deseas realizar otra consulta? (si/no): ")
        if continuar == "no":
            print("\n¡Gracias por visitarnos! Cuida tus dientes.")
```
